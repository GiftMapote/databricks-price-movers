```sql
USE CATALOG databricks_cata;
USE SCHEMA price_movers;

WITH latest_hour AS (
  SELECT MAX(hour_ts) AS max_hour
  FROM gold_hourly_movers
)
SELECT m.asset_id, m.vs_currency, m.hour_ts, m.avg_price, m.prev_avg_price, m.pct_change
FROM gold_hourly_movers m
JOIN latest_hour h
ON m.hour_ts = h.max_hour
ORDER BY m.pct_change DESC
```
