### Connectivity Test: Can this databricks reach CoinGecko
**WHY:** The project depends on calling an external public API

```python
import requests
```

```python
# CoinGecko provides a simple ping endpoint ro confirm the API is up and running.

ping_url = "https://api.coingecko.com/api/v3/ping"
  
try:
    # Make a small request with a short timeout
    r = requests.get(ping_url, timeout=20)

    # Print a small preview of the response to confirm it looks correct
    print("Status code:", r.status_code)
    print("Response preview: ", r.text[:200])

    # Fail the notebook if the response is not 200
    r.raise_for_status()

except requests.exceptions.Timeout:
    # If the request times out, fail the notebook
    print("Request timed out")
  
except requests.exceptions.RequestException as e:
    # If there is any other error, fail the notebook
    print("Request failed:", str(e))
```
